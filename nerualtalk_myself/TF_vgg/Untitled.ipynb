## 用keras实现异或(XOR)

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import numpy as np
```

设置每次训练使用的样本数(batch size)， 分类的数量(number of class)，以及训练的次数(epoch)

```python
batch_size = 1
num_classes = 1
epochs = 1000
```

训练数据&测试数据

```python
x_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
x_test = np.array([[0, 0], [1, 0]])
y_train = np.array([[0], [1], [1], [0]])
y_test = np.array([[0], [1]])
```

```python
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

定义一个序贯模型

```python
model = Sequential()
```

下面添加一个全连接层，输入层有2个特征，后面的隐藏层有2个单元，且隐藏层的激活函数为tanh
所有层的bias都是自动添加的

```python
model.add(Dense(2, activation='tanh', input_shape=(2,)))
```

最后的输出也是一个全连接层，激活函数为sigmoid

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

选择随机梯度下降方法(SGD)训练网络，设置学习率

```python
sgd = SGD(lr=0.7)
```

编译模型，使用二分类的交叉熵(binary_crossentropy)作为代价函数

```python
model.compile(loss='binary_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

训练模型，包含了前向传播、求参数的梯度和反向传播
使用验证数据进行验证

```python
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
```

查看结果

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print(model.predict_proba(x_train))
```
